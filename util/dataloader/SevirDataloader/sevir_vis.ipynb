```python
from sevir_dataset import SEVIRDataModule
```

```python
dm = SEVIRDataModule(data_dir='./', batch_size=4)
dm.setup()

dataloader = dm.train_dataloader()
```

```python
loader = iter(dataloader)
ir069, ir107, vis, vil = next(loader)

ir069.shape, ir107.shape, vis.shape, vil.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.imshow(vis[1, 0, :, :], cmap='gray')
plt.title('VIS Data')
plt.colorbar()

plt.subplot(2, 2, 2)
plt.imshow(ir069[1, 0, :, :], cmap='inferno')
plt.title('IR069 Data')
plt.colorbar()

plt.subplot(2, 2, 3)
plt.imshow(ir107[1, 0, :, :], cmap='inferno')
plt.title('IR107 Data')
plt.colorbar()

plt.subplot(2, 2, 4)
plt.imshow(vil[1, 0, :, :], cmap='viridis')
plt.title('VIL Data')
plt.colorbar()

plt.tight_layout()
plt.show()
```
