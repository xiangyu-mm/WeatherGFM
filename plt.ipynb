```python
from dataset.lowlevel_PromtGIP_dataloader import DatasetLowlevel_Train
from petrel_client.client import Client
import io
import numpy as np
client = Client(conf_path="~/petreloss.conf")

data_type=['vis','ir069','ir107','vil']
frame_id = 48
data_all = []
for itme in data_type:
    path = f"s3://sevir_pair/{itme}/2019/R19072901108250.npy"
    data = io.BytesIO(client.get(path))
    data = np.load(data)
    data = data[:, :, frame_id]
    if itme == "vis":
        print(data)
    data_all.append(data)

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))
plt.subplot(2, 4, 1)
plt.imshow(data_all[0], cmap='gray')
plt.title('VIS Data')
plt.colorbar()

plt.subplot(2, 4, 2)
plt.imshow(data_all[1], cmap='inferno')
plt.title('ir069 Data')
plt.colorbar()

plt.subplot(2, 4, 3)
plt.imshow(data_all[2], cmap='inferno')
plt.title('ir107 Data')
plt.colorbar()

plt.subplot(2, 4, 4)
plt.imshow(data_all[3], cmap='viridis')
plt.title('VIL Data')
plt.colorbar()

plt.subplot(2, 4, 5)
plt.imshow(data_all[0], cmap='gray')
plt.title('VIS Data')
plt.colorbar()

plt.subplot(2, 4, 6)
plt.imshow(data_all[1], cmap='inferno')
plt.title('ir069 Data')
plt.colorbar()

plt.subplot(2, 4, 7)
plt.imshow(data_all[2], cmap='inferno')
plt.title('ir107 Data')
plt.colorbar()

plt.subplot(2, 4, 8)
plt.imshow(data_all[3], cmap='viridis')
plt.title('VIL Data')
plt.colorbar()

plt.tight_layout()
plt.show()
```
